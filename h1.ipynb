```python
def is_interesting(amenity):
    boring = ['waste_basket', 'toilets', 'bench']
    if amenity in boring:
        return False
    else:
        return True
```

```python
def is_tourism(tags):
    if 'tourism' in tags:
        return True
    else:
        return False
```

```python
df['is_interesting'] = df['amenity'].apply(lambda x : is_interesting(x))
```

```python
df['is_tourism'] = df['tags'].apply(lambda x : is_tourism(x))
```

```python
df[(df['is_interesting'] == True) | (df['is_tourism'] == True)]
```
